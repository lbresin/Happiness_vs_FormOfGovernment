```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
# Democracy Index Data
# Source: https://en.wikipedia.org/wiki/Democracy_Index#cite_note-index2019-7
df_dem = pd.read_csv('democracy_index.csv', index_col=0)
df_dem = df_dem.set_index('Country')
df_dem
```

```python
# World Happiness Report 2019
# Source: https://worldhappiness.report/ed/2019/
df_happy = pd.read_csv('happiness_index.csv', index_col=0)
df_happy = df_happy.set_index('Country')
df_happy
```

```python
# Environmental Performance Index 2020
# Source: https://epi.yale.edu/downloads
df_epi = pd.read_csv('environmental_performance_index.csv', index_col=0)
df_epi = df_epi.set_index('country')
df_epi
```

```python
# Merge the two data sets
df = df_happy.merge(df_dem, how='inner',right_index=True, left_index=True)
df2 = df.merge(df_epi, how='inner', right_index=True, left_index=True)
```

```python
fig, ax = plt.subplots(figsize=(13/2.5, 13/2.5))


ax.scatter(df[df['Regime type'] == 'Full democracy']['Score_y'], df[df['Regime type'] == 'Full democracy'].Score_x, color='#0571b0', label='Full democracy', alpha=.7, edgecolors='none')
ax.scatter(df[df['Regime type'] == 'Flawed democracy']['Score_y'], df[df['Regime type'] == 'Flawed democracy'].Score_x, color='#92c5de', label='Flawed democracy', alpha=.7, edgecolors='none')
ax.scatter(df[df['Regime type'] == 'Hybrid regime']['Score_y'], df[df['Regime type'] == 'Hybrid regime'].Score_x, color='#f4a582', label='Hybrid regime', alpha=.7, edgecolors='none')
ax.scatter(df[df['Regime type'] == 'Authoritarian']['Score_y'], df[df['Regime type'] == 'Authoritarian'].Score_x, color='#ca0020', label='Authoritarian', alpha=.7, edgecolors='none')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
#ax.spines['bottom'].set_visible(False)
#ax.spines['left'].set_visible(False)

# label China
country='China'
ax.scatter(df.loc[country]['Score_y'], df.loc[country]['Score_x'],facecolors='none',edgecolors='black',  linewidth=2)
x_loc = df.loc[country]['Score_y']
y_loc = df.loc[country]['Score_x']
ax.plot([x_loc, x_loc+0.12], [y_loc, y_loc-0.35], linewidth=0.25, color='black')
ax.plot([x_loc+0.12, x_loc+0.15], [y_loc-0.35, y_loc-0.35], linewidth=0.25, color='black')
ax.text(x_loc+0.15, y_loc-0.40,country)

# label China
country='United States of America'
ax.scatter(df.loc[country]['Score_y'], df.loc[country]['Score_x'],facecolors='none',edgecolors='black',  linewidth=2)
x_loc = df.loc[country]['Score_y']
y_loc = df.loc[country]['Score_x']
ax.plot([x_loc, x_loc+0.55], [y_loc, y_loc-0.35], linewidth=0.25, color='black')
ax.plot([x_loc+0.55, x_loc+0.58], [y_loc-0.35, y_loc-0.35], linewidth=0.25, color='black')
ax.text(x_loc+0.58, y_loc-0.40,'USA')

# label China
country='Norway'
ax.scatter(df.loc[country]['Score_y'], df.loc[country]['Score_x'],facecolors='none',edgecolors='black',  linewidth=2)
x_loc = df.loc[country]['Score_y']
y_loc = df.loc[country]['Score_x']
ax.plot([x_loc, x_loc+0.12], [y_loc, y_loc-0.35], linewidth=0.25, color='black')
ax.plot([x_loc+0.12, x_loc+0.15], [y_loc-0.35, y_loc-0.35], linewidth=0.25, color='black')
ax.text(x_loc+0.15, y_loc-0.40,country)

# label China
country='Germany'
ax.scatter(df.loc[country]['Score_y'], df.loc[country]['Score_x'],facecolors='none',edgecolors='black',  linewidth=2)
x_loc = df.loc[country]['Score_y']
y_loc = df.loc[country]['Score_x']
ax.plot([x_loc, x_loc+0.12], [y_loc, y_loc-0.25], linewidth=0.25, color='black')
ax.plot([x_loc+0.12, x_loc+0.15], [y_loc-0.25, y_loc-0.25], linewidth=0.25, color='black')
ax.text(x_loc+.12, y_loc-0.30,country)

ax.set_ylabel('Happiness Index')
ax.set_xlabel('Democracy Index')
plt.legend(frameon=False)
plt.savefig('happiness_vs_democracy.png', dpi=300)
plt.savefig('happiness_vs_democracy.pdf', dpi=300)
```
